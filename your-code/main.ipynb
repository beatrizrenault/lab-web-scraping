````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url
url = 'https://twitter.com/'
````

````python
# your code here

def get_followers(user, count=100): #it's important, otherwise is forever
    results = api.followers(user, count=count)
    # print(results)
    followers = [pd.Series(follower._json) for follower in results]
    #print (len(followers))
    df = pd.DataFrame(followers)
    return df


run_ = get_followers(input('Type the @: ')) #the name of the user.
run_.head(10)
````
